# Explore the `breizhcrops` package
Explore the breizhcrops package, which is a package for the analysis of crop yield data in Brittany, France. Load the data, and see the differences between the packages and the loaded data from ELECTS. 

```python
import breizhcrops as bzh
dataset = bzh.BreizhCrops("frh04")
x, y, field_id = dataset[0]
```

```python
# explore 
index = 12900
x, y, field_id = dataset[index]
print("x.shape", x.shape)
print("y=", y)
print("field_id", field_id)
print("classes:", dataset.classes)
print("-----------------")
print("dataset.mapping:", dataset.mapping)
print("-----------------")
print("len(dataset):", len(dataset))
print("classname: ", dataset.classname)
print("-----------------")
print("dataset.index:", dataset.index)
```
